```python
import matplotlib.pyplot as plt
```

```python
months = ["Січ", "Лют", "Бер", "Кві", "Тра", "Чер"]

plan = [100, 120, 140, 160, 180, 200]
fact = [90, 130, 135, 170, 175, 210]
```

```python
plt.plot(months, plan, label="План")
plt.plot(months, fact, label="Факт")

plt.title("Планові та фактичні продажі")
plt.xlabel("Місяці")
plt.ylabel("Продажі")

plt.legend()

plt.show()
```

# Результати

У цьому завданні:

- Створено список місяців.
- Створено два списки значень: **План** та **Факт**.
- Побудовано один графік із двома лініями.
- Додано заголовок, підписи осей та легенду.
- Графік відображено на екрані.
