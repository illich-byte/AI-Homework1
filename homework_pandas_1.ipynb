```python
import pandas as pd
```

```python
books = {
    "Назва": [
        "Кобзар",
        "1984",
        "Майстер і Маргарита",
        "Гаррі Поттер",
        "Маленький принц",
        "Тореадори з Васюківки"
    ],
    "Автор": [
        "Тарас Шевченко",
        "Джордж Орвелл",
        "Михайло Булгаков",
        "Джоан Роулінг",
        "Антуан де Сент-Екзюпері",
        "Всеволод Нестайко"
    ],
    "Рік видання": [
        1840,
        1949,
        1967,
        1997,
        1943,
        2018
    ],
    "Ціна": [
        250,
        320,
        280,
        450,
        220,
        300
    ]
}

df = pd.DataFrame(books)
```

```python
print(df)
```

```python
average_price = df["Ціна"].mean()

print("Середня ціна книг:", average_price)
```

```python
books_after_2015 = df[df["Рік видання"] > 2015]

print("Книги, видані після 2015 року:")
print(books_after_2015)
```

```python
sorted_df = df.sort_values(by="Ціна")

print("Відсортований DataFrame:")
print(sorted_df)
```

- Виведено весь DataFrame.

- Знайдено книги, видані після 2015 року.
- Відсортовано DataFrame за ціною у порядку зростання.
