```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

df = pd.DataFrame({
    "date": pd.date_range("2025-01-01", periods=30),
    "users": np.random.randint(100, 500, 30),
    "sessions": np.random.randint(150, 700, 30),
    "revenue": np.random.randint(1000, 6000, 30)
})

print(df)
```

```python
correlation = df[["users", "sessions", "revenue"]].corr()

print("Кореляційна матриця:")
print(correlation)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["users"], df["sessions"])
plt.title("Users vs Sessions")
plt.xlabel("Users")
plt.ylabel("Sessions")
plt.grid(True)
plt.show()

plt.figure(figsize=(6,4))
plt.scatter(df["users"], df["revenue"])
plt.title("Users vs Revenue")
plt.xlabel("Users")
plt.ylabel("Revenue")
plt.grid(True)
plt.show()

plt.figure(figsize=(6,4))
plt.scatter(df["sessions"], df["revenue"])
plt.title("Sessions vs Revenue")
plt.xlabel("Sessions")
plt.ylabel("Revenue")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["date"], df["revenue"], marker="o")

plt.title("Revenue за датами")
plt.xlabel("Дата")
plt.ylabel("Revenue")

plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

plt.show()
```
