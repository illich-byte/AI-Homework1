```python
temperature = [2, 5, 4, 6, 7, 3, 1]
```

```python
average = sum(temperature) / len(temperature)
maximum = max(temperature)
minimum = min(temperature)
difference = maximum - minimum

print("Середнє:", average)
print("Максимум:", maximum)
print("Мінімум:", minimum)
print("Різниця:", difference)
```

# Результати

| Показник | Значення |
|----------|----------|
| Середнє значення | 4.0 |
| Максимальне значення | 7 |
| Мінімальне значення | 1 |
| Різниця між максимумом і мінімумом | 6 |
