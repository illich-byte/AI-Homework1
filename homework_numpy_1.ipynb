```python
import numpy as np

array = np.random.randint(0, 51, 20)

print("Вихідний масив:")
print(array)
```

```python
threshold = int(input("Введіть порогове значення: "))
```

```python
count = np.sum(array > threshold)

print("Кількість елементів більших за", threshold, "=", count)
```

```python
maximum = np.max(array)
position = np.argmax(array)

print("Максимальне значення:", maximum)
print("Позиція першої появи:", position)
```

```python
sorted_array = np.sort(array)[::-1]

print("Відсортований масив:")
print(sorted_array)
```

- Обчислено кількість елементів, більших за порогове значення.
- Знайдено максимальний елемент та позицію його першої появи.
